```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

cust_df = pd.read_csv('./train_santander.csv', encoding='latin-1')
print('dataset shape : ', cust_df.shape)
cust_df.sample(5)
```

```python
print(cust_df['TARGET'].value_counts())
unsatisfied_cnt = cust_df[cust_df['TARGET']==1].TARGET.count()
total_cnt = cust_df.TARGET.count()
print('unsatisfied  비율: {0:.2f}'.format(unsatisfied_cnt/total_cnt))

import matplotlib.pyplot as plt
plt.hist(cust_df.TARGET, color='r')
plt.xlabel("TARGET Distribution")
plt.ylabel("NUMBER")
```

```python
cust_df.describe()
```

```python
cust_df['var3'].replace(-999999, 2, inplace=True)
cust_df.drop('ID', axis=1, inplace=True)

X_features = cust_df.iloc[:, :-1]
y_label = cust_df.iloc[:, -1]
print('피처 데이터 shape ; {0}'.format(X_features.shape))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_features, y_label, test_size=0.2, random_state=0, stratify = y_label)

train_cunt = y_train.count()
test_cunt = y_test.count()

print(f'학습 세트 shape: {X_train.shape}, 테스트 세트 shape: {X_test.shape}')
print('학습 세트 레이블 값 분포 비율')
print(y_train.value_counts()/train_cunt)
print('\n테스트 데이터 세트 레이블 값 분포 비율')

print(y_test.value_counts()/test_cunt)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, roc_auc_score, f1_score, precision_score, recall_score
```

```python
dt_clf_unpruned = DecisionTreeClassifier(random_state=123)
dt_clf_pruned = DecisionTreeClassifier(max_depth=3, random_state=123)

rf_clf_100 = RandomForestClassifier(n_estimators=100, random_state=123)
rf_clf_500 = RandomForestClassifier(n_estimators=500, random_state=123)

dt_clf_unpruned.fit(X_train, y_train)
dt_clf_pruned.fit(X_train, y_train)
rf_clf_100.fit(X_train, y_train)
rf_clf_500.fit(X_train, y_train)

clf_names = ["Decision (unpruned)", "Decision (pruned)", "Random (n=100)", "Random (n=500)"]
clfs = [dt_clf_unpruned, dt_clf_pruned, rf_clf_100, rf_clf_500]

for clf_n, clf_e in zip(clf_names, clfs):
  print("{} - ACC : {:.4f} | Prec : {:.4f} | Recall : {:.4f} | F1 : {:.4f} | AUC : {:.4f}".format(clf_n,
                                                                                                  accuracy_score(y_test, clf_e.predict(X_test)),
                                                                                                  precision_score(y_test, clf_e.predict(X_test)),
                                                                                                  recall_score(y_test, clf_e.predict(X_test)),
                                                                                                  f1_score(y_test, clf_e.predict(X_test)),
                                                                                                  roc_auc_score(y_test,clf_e.predict(X_test))))
```

```python
dt_params = {'max_depth':[2,3,4,5,6,7]}
rf_params = {'n_estimators':[100,300,500]}

dt_clf = DecisionTreeClassifier(random_state=123)
rf_clf = RandomForestClassifier(random_state=123)


gridcv_dt = GridSearchCV(dt_clf, param_grid=dt_params, scoring="roc_auc", cv = 3)
gridcv_rf = GridSearchCV(rf_clf, param_grid=rf_params, scoring="roc_auc", cv = 3)

gridcv_dt.fit(X_train, y_train)
gridcv_rf.fit(X_train, y_train)

print('DecisionTree _ GridSearchCV 최적 파라미터 : ', gridcv_dt.best_params_)
print('RandomForest _ GridSearchCV 최적 파라미터 : ', gridcv_rf.best_params_)
```

```python
best_dt = DecisionTreeClassifier(max_depth=6, random_state=123)
best_rf = RandomForestClassifier(n_estimators=500, random_state=123)

best_dt.fit(X_train, y_train)
best_rf.fit(X_train, y_train)

for clf_n, clf_e in zip(["DT", "RF"], [best_dt, best_rf]):
  print("{} - ACC : {:.4f} | Prec : {:.4f} | Recall : {:.4f} | F1 : {:.4f} | AUC : {:.4f}".format(clf_n,
                                                                                                  accuracy_score(y_test, clf_e.predict(X_test)),
                                                                                                  precision_score(y_test, clf_e.predict(X_test)),
                                                                                                  recall_score(y_test, clf_e.predict(X_test)),
                                                                                                  f1_score(y_test, clf_e.predict(X_test)),
                                                                                                  roc_auc_score(y_test,clf_e.predict_proba(X_test)[:,1])))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

model = RandomForestClassifier(random_state=123, n_estimators=500)
model.fit(X_train, y_train)

ft_importance_values = model.feature_importances_

ft_series = pd.Series(ft_importance_values, index = X_train.columns)
ft_top10 = ft_series.sort_values(ascending=False)[:10]

plt.figure(figsize=(8,6))
plt.title('Feature Importance Top 20')
sns.barplot(x=ft_top10, y=ft_top10.index)
plt.show()
```
