```python
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"]
    },
    index=[0,1,2,3]
)
```

```python
left = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)
```

```python
right = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
  
    }
)
```

```python
result= pd.merge(left, right, on="key")
```

```python
print(result)
```

```python
left = pd.DataFrame(
    {
        "key1": ["K0", "K0", "K1", "K2"],
        "key2": ["K0", "K1", "K0", "K1"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)
```

```python
right = pd.DataFrame(
    {
        "key1": ["K0", "K1", "K1", "K2"],
        "key2": ["K0", "K0", "K0", "K0"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
  
    }
)
```

```python
result= pd.merge(left, right, how="left", on=["key1", "key2"])
print(result)
```

```python
result= pd.merge(left, right, how="right", on=["key1", "key2"])
print(result)
```

```python
result= pd.merge(left, right, how="outer", on=["key1", "key2"])
print(result)
```

```python
result= pd.merge(left, right, how="inner", on=["key1", "key2"])
print(result)
```

```python
result= pd.merge(left, right, how="cross")
print(result)
```

```python
import numpy as np
df = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
    }
)
```

```python
grouped = df.groupby("A")
grouped = df.groupby(["A", "B"])
```

```python
df
```

```python
df2 = df.set_index(["A", "B"])
grouped = df2.groupby(level=df2.index.names.difference(["B"]))
grouped.sum()
```

```python
grouped = df.groupby("A")
grouped.aggregate(np.sum)
grouped = df.groupby(["A", "B"])
grouped.aggregate(np.sum)
```

```python
grouped = df.groupby(["A", "B"], as_index=False)
grouped.aggregate(np.sum)
df.groupby("A", as_index=False).sum()
```
